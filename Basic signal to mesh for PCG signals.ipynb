```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
# Load a .wav file (Heart Sound or ECG stored in audio format)
fs, signal = wavfile.read("/content/a0001.wav")
```

```python
signal = signal[:20000]
# Normalize and downsample if needed
signal = signal / np.max(np.abs(signal))  # Normalize between -1 and 1
#signal = signal[::5]  # Downsample for better visualization

plt.plot(signal)
plt.title("Heart Sound Signal")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Convert time into an angular coordinate (Î¸)
theta = np.linspace(0, 2 * np.pi, len(signal))  # Wrap around a cylinder

# Cylinder radius
R = 1

# Convert to 3D coordinates
x = R * np.cos(theta)
y = R * np.sin(theta)
z = signal  # The signal determines height

# Plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, color='b', linewidth=2)

ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Signal Amplitude")
ax.set_title("Signal Embedded on a Cylinder")

plt.show()
```

```python
# Generate a sphere
phi = np.linspace(0, np.pi, len(signal))  # Latitude
theta = np.linspace(0, 2 * np.pi, len(signal))  # Longitude

# Define base sphere radius
R = 2

# Map signal onto the radius
r = R + 0.2 * signal  # Small perturbation by signal

# Convert to 3D coordinates
x = r * np.sin(phi) * np.cos(theta)
y = r * np.sin(phi) * np.sin(theta)
z = r * np.cos(phi)

# Plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, color='r', linewidth=2)

ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Signal Amplitude")
ax.set_title("Signal Embedded on a Sphere")

plt.show()
```
